```python
import matplotlib.pyplot as plt

def plot_scenario(power_links, df_input):

  pos_df = df_input[['node_code', 'x(m)', 'y(m)', 'z(m)', 'primary_channel']].copy()
  pos_df['node_code'] = pos_df['node_code'].str.strip()  # Clean up node codes

  # Set up plot
  fig, ax = plt.subplots(figsize=(8, 6))

  # Map node code to position for lookup
  node_positions = {
      row['node_code']: (row['x(m)'], row['y(m)'])
      for _, row in pos_df.iterrows()
  }

  # Plot nodes
  for _, row in pos_df.iterrows():
      node_code = row['node_code']
      x, y = row['x(m)'], row['y(m)']
      color = plt.cm.tab10(row['primary_channel'] % 10)
      marker = 's' if node_code in ['STA_A1', 'STA_B1'] else 'o' # ap circles and sta squares

      ax.scatter(x, y, c=[color], marker=marker, s=100, edgecolor='black')
      ax.text(x, y, node_code, ha='right', va='bottom', fontsize=9)

  # Plot arrows for power sensed relationships
  for (receiver, transmitter), power in power_links.items():
      if receiver in node_positions and transmitter in node_positions:
          x1, y1 = node_positions[receiver]
          x2, y2 = node_positions[transmitter]

          # Draw arrow from receiver to transmitter
          ax.annotate(
              '', xy=(x2, y2), xytext=(x1, y1),
              arrowprops=dict(arrowstyle='->', color='red', lw=1.5, alpha=0.7)
          )

          # Annotate power at midpoint
          mid_x, mid_y = (x1 + x2) / 2, (y1 + y2) / 2
          ax.text(mid_x, mid_y, f'{power:.1f} dBm',
                  fontsize=8, color='red', ha='center', va='center')

  # Final touches
  ax.set_xlabel('X position (m)')
  ax.set_ylabel('Y position (m)')
  ax.set_title('Node Positions and Power Sensed Links')
  plt.grid(True, linestyle='--', alpha=0.5)
  plt.tight_layout()
  plt.show()

  # falta plot del canal? pero todos son 0
```

```python
import matplotlib.pyplot as plt
import glob

def plot_avg_throughput(log_files):

  # Prepare the figure
  plt.figure(figsize=(10, 6))

  # Process each file
  for log_file in log_files:
      times = []
      throughputs = []
      with open(log_file, 'r') as f:
          lines = f.readlines()
          for i in range(len(lines)):
              if "RequestInformationToAp" in lines[i]:
                  # Extract time
                  time = float(lines[i].split(";")[0])
              if "Average throughput" in lines[i]:
                  # Extract throughput
                  parts = lines[i].split("=")
                  if len(parts) > 1:
                      throughput = float(parts[1].split()[0])
                      times.append(time)
                      throughputs.append(throughput)

      ap_name = log_file.split("/")[-1].split(".")[0]
      plt.plot(times, throughputs, label=f"{ap_name}")

  # Plot formatting
  plt.xlabel("Time (s)")
  plt.ylabel("Average Throughput (Mbps)")
  plt.title("Average Throughput Over Time per AP")
  plt.legend()
  plt.grid(True)
  plt.tight_layout()
  plt.show()
```

### Exercise 5:

Run a simulation to test your action-selection strategy. Compare it to ɛ-greedy (Exercise 3) by plotting the temporary throughput obtained by each of the mechanisms.

```python
import os
import glob
import matplotlib.pyplot as plt

# 1) Define your two strategies and grab *all* the AP‐files into one dict
patterns = {
    'AgentsML': os.path.join(mypath, 'output', 'logs_output_AAX-LAB1-AGENTSML*.txt'),
    'Agents3':  os.path.join(mypath, 'output', 'logs_output_AAX-LAB1-AGENTS-3_*.txt'),
}

series = {}  # label -> (times, throughputs)

for strat_name, pattern in patterns.items():
    for log_file in glob.glob(pattern):
        times, ths = [], []
        with open(log_file) as f:
            for line in f:
                if "RequestInformationToAp" in line:
                    t = float(line.split(";")[0])
                elif "Average throughput" in line:
                    parts = line.split("=")
                    if len(parts) > 1:
                        ths.append(float(parts[1].split()[0]))
                        times.append(t)
        ap = os.path.basename(log_file).rsplit(".",1)[0]
        label = f"{strat_name} | {ap}"
        series[label] = (times, ths)

# 2) Make a 5×1 grid: 1 big up top + 4 small below
fig, axes = plt.subplots(
    nrows=5,
    ncols=1,
    figsize=(10, 12),
    sharex=True,
    gridspec_kw={'height_ratios': [2, 1, 1, 1, 1]}
)

# Top panel: all together
ax0 = axes[0]
for lbl, (t, th) in series.items():
    ax0.plot(t, th, label=lbl)
ax0.set_title("Combined: AgentsML vs. Agents3")
ax0.set_ylabel("Throughput (Mbps)")
ax0.legend(fontsize='small', loc='best')
ax0.grid(True)

# The four individual panels
for ax, (lbl, (t, th)) in zip(axes[1:], series.items()):
    ax.plot(t, th)
    ax.set_ylabel("Mbps")
    ax.set_title(lbl, fontsize='small')
    ax.grid(True)

# X‐axis label on the bottom plot only
axes[-1].set_xlabel("Time (s)")

plt.tight_layout()
plt.show()
```

What we can see clearly here is that our implementation seems to provide a more stable average throughput of 40Mbps while the e-greedy is oscilating between 20~40Mbps

#### Plots

```python
import glob
import os
import matplotlib.pyplot as plt

sims = range(1, 6)
approaches = {
    'ML4NET':    "logs_output_ML4NET-BSS-{}.txt",
    'EGREEDY':   "logs_output_EGREEDY-BSS-{}.txt"
}

fig, axes = plt.subplots(nrows=5,
                         ncols=2,
                         figsize=(12, 15),
                         sharex=True,
                         sharey=True,
                         gridspec_kw={'hspace': 0.3, 'wspace': 0.2})

for i, sim in enumerate(sims):
    for j, (name, pattern_tmpl) in enumerate(approaches.items()):
        ax = axes[i, j]
        pattern = os.path.join(mypath,
                               'output',
                               pattern_tmpl.format(sim).replace('.txt', '*.txt'))
        for log_file in glob.glob(pattern):
            times, ths = [], []
            with open(log_file) as f:
                for line in f:
                    if "RequestInformationToAp" in line:
                        t = float(line.split(";")[0])
                    elif "Average throughput" in line:
                        parts = line.split("=")
                        if len(parts) > 1:
                            times.append(t)
                            ths.append(float(parts[1].split()[0]))
            ap = os.path.basename(log_file).split('.')[0]
            ax.plot(times, ths, label=ap, alpha=0.7)

        ax.set_title(f"{name} sim {sim}", fontsize='small')
        ax.grid(True)
        if i == 4:
            ax.set_xlabel("Time (s)")
        if j == 0:
            ax.set_ylabel("Throughput (Mbps)")
        if i == 0 and j == 1:
            ax.legend(fontsize='x-small', loc='upper right')

plt.suptitle("Avg Throughput: ML4NET vs EGREEDY, Sims 1–5", y=0.92)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

ML4NET delivers both higher average throughput and far lower variance we quickly get a sustained ~40 Mbps on every AP.

EGREEDY occasionally spikes to 40 Mbps but then crashes back down repeatedly, making it unpredictable and much harder to provision around.

If we are valuing steady, predictable performance, ML4NET is clearly the better choice here.
